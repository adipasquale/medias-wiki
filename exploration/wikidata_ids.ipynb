```python
import pandas as pd
import requests
```

```python
def download_file(url, filename):
  with open(filename, "wb") as f:
    f.write(requests.get(url).content)
```

```python
download_file("https://raw.githubusercontent.com/mdiplo/Medias_francais/master/medias_francais.tsv", "current.tsv")
download_file("https://raw.githubusercontent.com/johanricher/Medias_francais/wikidata/medias_francais.tsv", "old_with_wikidata.tsv")
```

```python
df_current = pd.read_csv("current.tsv", sep="\t")
df_old_with_wikidata = pd.read_csv("old_with_wikidata.tsv", sep="\t")
```

```python
df_diff = pd.merge(df_current[['id', 'nom']], df_old_with_wikidata[['id', 'nom']], on='id', suffixes=('_current', '_old_with_wikidata'))
df_diff[df_diff['nom_current'] != df_diff['nom_old_with_wikidata']]
```

```python
df_current_with_wikidata = pd.merge(df_current, df_old_with_wikidata[["id", "wikidata_id"]], on='id', how="left")
df_current_with_wikidata
```

```python
df_current_with_wikidata.to_csv("current_with_wikidata.tsv", sep="\t", index=False, float_format='%.0f')
```
